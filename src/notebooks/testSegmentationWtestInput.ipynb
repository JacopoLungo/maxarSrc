```python
import leafmap
from samgeo import tms_to_geotiff
from samgeo.text_sam import LangSAM
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image = '/home/vaschetti/maxarSrc/exampleData/10300100C2D81700-visual_small.tif'
# Load the PNG image
img = mpimg.imread(image)

# Display the image
plt.imshow(img)
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
sam = LangSAM()
```

```python
text_prompt = "house"
sam.predict(image, text_prompt, box_threshold=0.20, text_threshold=0.24)
sam.show_anns(
    cmap='Greens',
    box_color='red',
    title='Automatic Segmentation of Trees',
    blend=True,
)
```
