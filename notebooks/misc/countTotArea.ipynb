```python
import glob
import os
from pathlib import Path
import geopandas as gpd
import pandas as pd
```

```python
root = Path('/home/vaschetti/maxarSrc/metadata/from_github_maxar_metadata/datasets')
```

```python
all_events = [d for d in os.listdir(root) if os.path.isdir(os.path.join(root, d))]
all_events
```

```python
tot_images = 0
diz_df = {'event_name': [], 'num_img': []}
for event_name in all_events:
    event_imgs = 0
    event_path = root / event_name
    all_geojson = list(event_path.glob('*.geojson'))
    for geojson in all_geojson:
        if '_union' not in str(geojson):
            #print(geojson)
            gdf = gpd.read_file(geojson)
            event_imgs += len(gdf)
            #tile_area = gdf['tile:data_area'].sum()
            #event_area += tile_area
            #print(gdf['tile:data_area'].sum())
    diz_df['event_name'].append(event_name)
    diz_df['num_img'].append(event_imgs)
    tot_images += event_imgs
    
print(tot_images)
df = pd.DataFrame(diz_df)
df.to_csv('/home/vaschetti/maxarSrc/output/num_imgs.csv', index=False)
```

```python
df = pd.read_csv('/home/vaschetti/maxarSrc/output/areakm2.csv')
```

```python
tot_area = 0
diz_df = {'event_name': [], 'area_km2': []}
for event_name in all_events:
    event_area = 0
    event_path = root / event_name
    #print(event_path)
    all_geojson = list(event_path.glob('*.geojson'))
    #print(all_geojson)
    for geojson in all_geojson:
        if '_union' not in str(geojson):
            #print(geojson)
            gdf = gpd.read_file(geojson)
            tile_area = gdf['tile:data_area'].sum()
            event_area += tile_area
            #print(gdf['tile:data_area'].sum())
    diz_df['event_name'].append(event_name)
    diz_df['area_km2'].append(event_area)
    tot_area += event_area
    
print(tot_area)
df = pd.DataFrame(diz_df)
df.to_csv('/home/vaschetti/maxarSrc/output/areakm2.csv', index=False)
```
