```python
#Get mosiacs names given event name
data_root = Path('/mnt/data2/vaschetti_data/maxar')
event_name = 'Gambia-flooding-8-11-2022'
all_mosaic = []
when = 'pre'
for mosaic_name in glob.glob('**/*', root_dir=data_root/event_name):
    all_mosaic.append(mosaic_name.split('/')[1])
print(all_mosaic)
```

```python
# given a mosaic name get extention (top left corner and bottom right corner)
def get_mosaic_extension(event_name, mosaic_name, path_mosaic_metatada = './metadata/from_github_maxar_metadata/datasets', return_lon_lat = False):
    path_mosaic_metatada = Path(path_mosaic_metatada)
    file_name = mosaic_name+'.geojson'
    geojson_path = path_mosaic_metatada / event_name / file_name
    gdf = gpd.read_file(geojson_path)

    minx = sys.maxsize
    miny = sys.maxsize
    maxx = 0
    maxy = 0

    for _, row in gdf.iterrows():

        tmp_minx, tmp_miny, tmp_maxx, tmp_maxy = [float(el) for el in row['proj:bbox'].split(',')]
        if tmp_minx < minx:
            minx = tmp_minx
        if tmp_miny < miny:
            miny = tmp_miny
        if tmp_maxx > maxx:
            maxx = tmp_maxx
        if tmp_maxy > maxy:
            maxy = tmp_maxy
    
    if return_lon_lat:
        source_crs = gdf['proj:epsg'].values[0]
        target_crs = pyproj.CRS('EPSG:4326')
        transformer = pyproj.Transformer.from_crs(source_crs, target_crs)
        bott_left_lat, bott_left_lon = transformer.transform(minx, miny)
        top_right_lat, top_right_lon = transformer.transform(maxx, maxy)
        return bott_left_lon, bott_left_lat, top_right_lon, top_right_lat
    
    return minx, miny, maxx, maxy

mosaic_name = all_mosaic[0]
minx, miny, maxx, maxy = get_mosaic_extension(event_name, mosaic_name)
bott_left_lon, bott_left_lat, top_right_lon, top_right_lat = get_mosaic_extension(event_name, mosaic_name, return_lon_lat=True)


print(minx, ',', miny,'---',  maxx, ',', maxy)
print(bott_left_lon, ',', bott_left_lat,'---',  top_right_lon, ',', top_right_lat)
```

```python
from pyquadkey2 import quadkey
def intersecting_qks(bott_left_lon, bott_left_lat, top_right_lon, top_right_lat, min_level=7, max_level=9):
    qk_bott_left = quadkey.from_geo((bott_left_lat, bott_left_lon), level=max_level) #lat, lon
    qk_top_right = quadkey.from_geo((top_right_lat, top_right_lon), level=max_level) #lat, lon
    hits = qk_bott_left.difference(qk_top_right)
    candidate_hits = set()

    for hit in hits:
        #tmp_candidate = set()
        current_qk = hit
        for _ in range(min_level, max_level):
            current_qk = current_qk.parent()
            candidate_hits.add(current_qk)
    hits.extend(candidate_hits)
    return [int(str(hit)) for hit in hits]
```
